# Flipping Images

```python
image = Image.open('img.jpg')
plt.imshow(image)
```

```python
img_flip = ImageOps.flip(image)
plt.imshow(img_flip)
```

```python
img_mirror = ImageOps.mirror(image)
plt.imshow(img_mirror)
```

```python
  flip = {"FLIP_LEFT_RIGHT": Image.FLIP_LEFT_RIGHT,
        "FLIP_TOP_BOTTOM": Image.FLIP_TOP_BOTTOM,
}
```

```python
for key, values in flip.items():
    plt.subplot(1,2,1)
    plt.imshow(image)
    plt.title("orignal")
    plt.subplot(1,2,2)
    plt.imshow(image.transpose(values))
    plt.title(key)
    plt.show()
```

```python
  rotate = {"ROTATE_90": Image.ROTATE_90,
        "ROTATE_180": Image.ROTATE_180,
        "ROTATE_270": Image.ROTATE_270,}
```
